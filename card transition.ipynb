```python
import scipy 
from scipy import stats
from scipy.stats import zscore
```

```python
from scipy.stats.mstats import winsorize
```

```python
zscore_distance_from_home = zscore(df["distance_from_home"])
zscore_distance_from_home
```

```python
for i in range(1,4):
    print("Value: {}".format(i))
    print("Outliers: {}".format(len(np.where(zscore_distance_from_home > i)[0])))
```

```python
df[zscore_distance_from_home<2][["distance_from_home"]].distance_from_home.min()
```

```python
df[zscore_distance_from_home<2][["distance_from_home"]].distance_from_home.max()
```

```python
w_percentile_distance_from_home = (stats.percentileofscore(df["distance_from_home"],157.406634599154)/100)
w_percentile_distance_from_home
```

```python
1-w_percentile_distance_from_home
```

```python
dist_from_home_w = winsorize(df.distance_from_home,(0,(1-w_percentile_distance_from_home)))
```

```python
plt.boxplot(dist_from_home_w)
plt.show()
```
